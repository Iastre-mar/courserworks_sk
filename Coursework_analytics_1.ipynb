```python
main2 = main[main['is_homework'] == True]
main2
#Выделил из таблицы значения, связанные с домашней работой
```

```python
first_hw_df = main2[['title', 'module_number', 'lesson_number', 'student_id', 'start_date']]
first_module = pd.DataFrame(first_hw_df[['title', 'module_number']].groupby('title')['module_number'].unique())
first_module['module_number'] = first_module['module_number'].apply(min)
first_lesson = first_module.reset_index().merge(first_hw_df)
first_lesson_test = first_lesson.drop(columns=['student_id', 'start_date']).groupby(['title', 'module_number'])['lesson_number'].unique()
#В кажом первом модуле по одному уроку с домашним заданием
test_one_student_one_first_try = first_lesson.groupby(['title', 'module_number','lesson_number' ])['student_id'].unique().apply(len).sum()
#Каждый студент начинал первую домашнюю работу один раз
min_first = pd.DataFrame(first_lesson.groupby(['title', 'module_number','lesson_number'])['start_date'].agg('unique').apply(min))
max_first = pd.DataFrame(first_lesson.groupby(['title', 'module_number','lesson_number'])['start_date'].agg('unique').apply(max))
#Нашел даты первого и последнего приростов студентов на каждом из курсов для создания интервала
first_lesson
```

```python
def inter(x, freq = 'M'):
    'Возвращает промежуток дат между первыми 2 значениями кортежа/массива, можно задать тип промежутка'
    try:
        end = pd.date_range(start = x[1], freq=freq, periods = 1,  normalize = True)[0]
        return pd.date_range(start = x[0], end = end, freq=freq, normalize = True)
    except:
        return pd.date_range(start = x[0], end = x[0], freq=freq, normalize = True)
    
interval = min_first.reset_index().rename(columns={'start_date':'start'}).merge(max_first.reset_index().rename(columns={'start_date':'end'})).drop(columns=['module_number', 'lesson_number'])
interval['range'] = list(zip(interval['start'], interval['end']))
interval['interval'] = interval['range'].apply(inter)
#Нашел интервал посткплений студентов для каждого курса
```

```python
dct = interval[['title','interval']]
sol_dict_1 = dct.set_index('title').to_dict().get('interval')
# Создал словарь с датами поступления студентов для каждого курса
```

```python
def sravnenie(x):
    'Сравнивает 2 даты в кортеже'
    for y in x[0]:
        if y >= x[1]:
            return y

time_normalizer = pd.Timedelta('1 days')
first_lesson['prirost'] = (pd.to_datetime(first_lesson['start_date'])) - time_normalizer
dct2 = first_lesson[['title', 'prirost']]
dct3 = dct.merge(dct2).set_index('title')
dct3['sravn'] = list(zip(dct3['interval'], dct3['prirost']))
dct3['sravn2'] = dct3['sravn'].apply(sravnenie)
dct4 = dct3.reset_index()[['title', 'sravn2']]
dct4['count'] = 1
dct4 = pd.DataFrame(dct4.groupby(['title', 'sravn2'])['count'].agg('sum'))
sol_dict_2 = dct4.to_dict().get('count')
# Создал словарь с распределенными по месяцам значениями поступающих
```

```python
fin_dict = {}
# создал финальный словарь, который послужит для создания финального датафрейма, который ответиит на поставленную задачу
```

```python
for keys in sol_dict_2:
    for titles in sol_dict_1:
        if keys[0] == titles:
            for elem in sol_dict_1.get(titles):
                if elem == keys[1]:
                    elem = {elem :sol_dict_2.get(keys)}
                    if fin_dict.get(titles) == None:
                        fin_dict[titles] = elem
                        break
                    else:
                        fin_dict.get(titles).update(elem)
                        break
# добавил в финальный словарь        
```

```python
for titles in sol_dict_1:
    for titles2 in fin_dict:
        if titles == titles2:
            for values in sol_dict_1.get(titles):
                if values not in fin_dict.get(titles2):
                    fin_dict.get(titles2).update({values : 0})
# добавил в финальный словарь месяца в которые студенты не поступали
```

```python
def pan_gen_dict(x):
    'Превращает диапазон дат в словарь где каждой дате присвоено значение 1'
    dct = {}
    for z in x:
        dct.update({z:1})
        x = dct
    return dct

def pan_gen_dict_zero(x):
    'Превращает диапазон дат в словарь где каждой дате присвоено значение 0'
    dct = {}
    for z in x:
        dct.update({z:0})
        x = dct
    return dct

dict_few = few_months_df.set_index(['title', 'module_number', 'lesson_number', 'student_id'])
dict_few['kostil'] = dict_few['inter'].apply(pan_gen_dict)
# Дал каждой дате в интервале дат, в ходе которых выполнялось домашнее задание, значение 1, необхлдимо для подсчета
interval2_dict = interval2['interval']
#Начинаю создавать словарь со значениями всех дат для каждого курса, необходимо для подсчета
    
pre_fin1 = interval2_dict.apply(pan_gen_dict_zero)
pre_fin1 = pd.DataFrame(pre_fin1).reset_index().set_index('title')['interval'].to_dict()
# словарь где ключами являются названия курсов и даты создан
dict_few = dict_few['kostil']
grouped = pd.DataFrame(pd.DataFrame(dict_few)).to_dict().get('kostil')
#Словарь с значениями для каждой работы, содержит много ключей для сохранения всех значений
for x in grouped:
    for y in pre_fin1:
        if x[0] == y:
            for z in grouped.get(x):
                try:
                    pre_fin1[y][z] += grouped[x][z] 
                except:
                    pre_fin1[y][z] = grouped[x][z]
# Словарь для создания датафрейма, с данными для каждого месяца каждого курса о выполненных за несколько месяцах работах

fin_df2 = pd.DataFrame(pd.DataFrame.from_dict(pre_fin1, orient = 'index').stack()).reset_index().rename(columns = {'level_0':'title', 'level_1':'start_date', 0:'count2'}).set_index(['title', 'start_date'])
fin_df2 = fin_df2.reset_index().merge(one_month_df.reset_index(), how='outer').fillna(0)
fin_df2['count'] = fin_df2['count1'] + fin_df2['count2']
fin_df2 = fin_df2.drop(columns = ['count1', 'count2']).set_index(['title', 'start_date']).sort_index(level = ['title', 'start_date'])
# Соединил значения о прогрессах студентов в один датафрейм
fin_df2
```
